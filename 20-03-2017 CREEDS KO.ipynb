```python
# Clean/ Organize/ Standardize output from gene overlap comparisons.
```

```python
import re

# List of words to delete
delete = ['size','#','\n','Index',]

# Data to keep 
words = ['listA', 'listB','Intersection','Overlapping p-value', 'Odds ratio','Union','Jaccard Index']


rep = re.compile(r'|'.join(delete))
keep = re.compile(r'|'.join(words))

data = []
with open('/Users/Carol/Desktop/arquivoNovo/KO/overlapKOdn.txt') as f1:
    for line in f1:
        if keep.search(line):
            # replace all unwanted substrings
            line = rep.sub("", line.lstrip())     
            line=line.split('=')
#             print(line)
            data.append(line)


df = pd.DataFrame(data)

df.columns = ['ca', 'value']
df['value'] = df.value.str.split(',').str.get(0)  # split value column at the , and que left side of string


df.head()
```

```python
# Coisas que podem ser uteis
listA = df[df['ca'].str.contains("listA")]
listB = df[df['ca'].str.contains("listB")]
intersection = df[df['ca'].str.contains("Intersection")]
union = df[df['ca'].str.contains("Union")]
pval= df[df['ca'].str.contains("Overlapping p-value")]
odds= df[df['ca'].str.contains("Odds ratio")]
#jaccard= df[df['ca'].str.contains("Jaccard Index")]
```
